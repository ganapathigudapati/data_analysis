```python
#plot the scores of two students againest 5 five subjects
#subjects as x axis, and marks as y axis
std1=[56,78,92,95,78]
std2=[77,89,91,23,45]
sub=["math","sci","eng","tel","soc"]
plt.plot(sub,std1,label="student1",color="b")
plt.plot(sub,std2,label="student2",color="y")
plt.title("marks of student")
plt.legend()
```

```python
#plot the scores of two students againest 5 five subjects
#subjects as x axis, and marks as y axis
#creating sub plots
std1=[56,78,92,95,78]
std2=[77,89,91,23,45]
sub=["math","sci","eng","tel","soc"]
plt.subplot(2,2,1)
plt.plot(sub,std1,label="student1",color="b")
plt.legend()
plt.subplot(2,2,2)
plt.plot(sub,std2,label="student2",color="y")
plt.legend()

```

```python
#create sub plots(bar) showing the profits of two companies
A=np.array([230,560,780,127,128])
B=np.array([200,160,270,127,400])
print("revenues of company A",A)
print("revenues of company B",B)
year=np.array(["19-20","20-21","21-22","22-23"])
proa=np.diff(A)
print("profits of a",proa)
prob=np.diff(B)
print("profits of b",prob)
plt.subplot(2,2,1)
plt.bar(year,proa,label="profitA",color="k")
plt.legend(loc="best")
plt.subplot(2,2,2)
plt.bar(year,prob,label="profitB",color="y")
plt.legend(loc="best")
```
