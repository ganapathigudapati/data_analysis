```python
B=np.arange(1,33).reshape(4,8)
print(B)
```

```python
b=np.arange(1,33).reshape(4,8)
print(b)
```

```python
np.hsplit(B,4)      #SPLITS IN TO REQUIRED NUMBER OF PARTS HORIZONTALLY
```

```python
np.hsplit(B,(2,6))     #splits first two columns and 6 columns(from starting) and remaining
```

```python
A=[np.pi/4,np.pi/3,np.pi/2,np.pi]
print(A)
```

```python
k=np.rad2deg(A)    #radian to degree conversion
print(k)
```

```python
o=10*np.random.rand(20).reshape(5,4)
print(o)
np.argmax(o)
```

```python
print(np.argmax(o,axis=0))   #gives index of max value along columns
print(np.argmin(o,axis=0))    #GIVE INDEX OF MIN VALUE ALONG COLUMNS
```

print(np.argmax(o,axis=1))    #gives index of max value along rows
