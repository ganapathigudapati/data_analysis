```python
#importing
import numpy as np
```

```python
#ARRAYS
#creating 1d array
A=np.array([2,3,4,5])
P=np.array([6,7,8,9])
print(type(A))
print(A)
print(A.ndim)
```

```python
#creating two dimensional array
B=np.array([[2,3,4],[7,8,9]])
print(B)
G=np.array([[2,3,4],[7,8,9]])
print(B.ndim)
```

```python
#creating 3d array--rows,columns,groups
C=np.array([[[2,3,4],[6,7,8]],[[6,9,0],[2,3,1]]])
print(C)
print(C.ndim)
```

```python
#ones-2d array
D=np.ones((3,2))
print(D)
```

```python
#ones--3d array-groups,rows,columns
J=np.ones((2,3,2))
print(J)
```

```python
#zeroes-2darray
E=np.zeros((2,3))
print(E)
```

```python
#eyes-2darray
F=np.eye(4,4)
print(F)
```

```python
#arange
L=np.arange(1,31,1)
print(L)
```

```python
#arange with reshape
L=np.arange(3,31,3).reshape(5,2)
print(L)
```

```python
#arange
K=np.arange(5,1001,5)
print(K)
```

```python
#linspace
V=np.linspace(12,24,10)
print(V)
```

```python
#adding one dimensional arrays
print(A+P)
```

```python
#adding(adding two rows in together) row wise(axis=1)
I=np.sum((B,G),axis=1)
print(I)
```

```python
#adding column wise(axis=0)
I=np.sum((B,G),axis=0)
print(I)
```

```python
N=np.array([[1,1],[0,1]])
S=np.array([[2,0],[3,4]])
print(N*S)
print(N@S)
```

```python
b=np.array([25,289,361,81])
for i in b:
  print(np.sqrt(i),"square is",i)
```

```python
#joining two matrices
a=np.array([34,35,36,37,38,39])
a=a.reshape(2,3)
b=np.array([4,5,6,7,8,9])
b.resize(2,3)
print(np.vstack((a,b)))#columns(join columns wise)
print(np.hstack((a,b)))#rows(join rows wise)
```

```python
a=np.arange(30).reshape(2,3,5)
print(a)
print("\n")
print("output")
print(np.dstack(a))
#number of rows becomes number of groups
#columns becomes rows
#group becomes columns
```
